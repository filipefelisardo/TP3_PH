```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados da tabela
dados = {
    'Classe': list(range(1, 19)),
    'Caudal': [15, 20, 30, 40, 50, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 300, 320],
    'Área_regavel_Max': [2.74, 3.66, 5.49, 7.32, 10.97, 14.63, 19.25, 23.10, 26.95, 30.81, 34.66, 45.73, 50.30, 54.87, 59.45, 64.02, 68.59, 73.16]
}

# Criar um DataFrame
df = pd.DataFrame(dados)

# Criar o gráfico de linhas
plt.figure(figsize=(8, 6))
plt.step(df['Área_regavel_Max'], df['Caudal'], where='post', color='navy', linewidth=2)

# Configuração do gráfico
plt.xlabel('Área regável (ha)')
plt.ylabel('Caudal (m³/h)')
plt.grid(True)
plt.show()
```
